```python
#Preprocessing on a dataset-1
import pandas as pd

data = {
    "Roll No": [45,47,54,55,58],
    "Name": ["Drishti", "Falak", "Hiranya", "Jahnvi", "Jayvee"],
    "Age": [18, "-", 19, 21, "20"],
    "Marks": [ 85,75,89,92,90],
    "Department": ["CSE", "IT", "ENTC", "cse", "CSE"],
    "Admission_Date": ["01-06-2023", "03-06-2023", "05-06-2023", "07-06-2023", "10/06/23"]
    }
df = pd.DataFrame(data)

# Save as CSV
df.to_csv("student_data.csv", index= False)
```

```python
import pandas as pd
import numpy as np

#Step 1: Load the Dataset
df= pd.read_csv("student_data.csv")

print("Original Data: ")
print(df)
```

```python
df.replace("-", np.nan, inplace=True)
df
```

```python
#Step 2: Replace "-" with NaN to mark missing values

df.replace("-", np.nan, inplace=True)

#Convert columns to numeric where needed

df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Marks"] = pd.to_numeric(df["Marks"], errors="coerce")


'''When converting data (like strings to datatime),
if some values are invalid or cannot be converted,
instead of throwing an error, it will:
Replace those invalid values with NaT(Not a Time)'''

df
```

```python
#Step 3: Check missing values
print("\nMissing Values: ")
print(df.isnull().sum())
```

```python
#Fill missing age with mean

df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
df["Marks"].fillna(df["Marks"].median(), inplace=True)
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
#Step 5: Fix inconsistent text formatting
df["Department"] = df["Department"].str.upper()
df
```

```python
#Step 6: Convert Admission_Date to standard datatime format
df["Admission_Date"] = pd.to_datetime(df["Admission_Date"], errors='coerce')
```

```python
df
```

```python
#Step 7: Show cleaned data
print("\nCleaned data: ")
df
```

```python
#Step 8: Save cleaned data
df.to_csv("cleaned_student_data.csv", index=False)
print("\nCleaned CSV file saved as cleaned_student_data.csv")
```

```python
#Working with missing values on Dataset-2
df=pd.read_csv('/content/Cars93.csv')
df
```

```python
df.isna().sum() #Attributes are displayed
```

```python
#Imputing the missing value
#1. Luggage room
df['Luggage.room']=df['Luggage.room'].fillna(value=df['Luggage.room'].mean())
df
```

```python
#2. Rear.seat.room
df['Rear.seat.room']=df['Rear.seat.room'].fillna(value=df['Rear.seat.room'].mean())
df
```

```python
#3. Airbages
df['AirBags']=df['AirBags'].fillna(value=df['AirBags'].mode().values[0])
df
```

```python
#Save cleaned dataset
df.to_csv("cleaned_Cars93.csv", index=False)
```
